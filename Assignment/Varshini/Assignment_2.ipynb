# Finding The Outliers

The outliers are found using the Box Plot which is already performed in the Univariate Analysis.

# Replacing the Outliers

```python
CreditsMedian=df.loc[df['CreditScore']<400,'CreditScore'].median()
ProdMedian=df.loc[df['NumOfProducts']>=3.5,'NumOfProducts'].median()
df.loc[df.CreditScore<400,'CreditScore']=np.nan
df.fillna(CreditsMedian,inplace=True)
df.loc[df.NumOfProducts>3,'NumOfProducts']=np.nan
df.fillna(ProdMedian,inplace=True)
df.head()
```

# Checking For Categorial Columns and Performing encoding

Geography and Gender are the two Categorial columns in the dataset so we can perform the Label Encoding in these two columns.

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df.Geography=le.fit_transform(df.Geography)
```
